```python
def get_sta(stim, spks, prespk, postspk):
    allregion = np.zeros((len(spks), 
                          prespk+postspk))
    for i, spk in enumerate(spks):
        if (spk > prespk 
            and len(stim) - postspk > spk):
            allregion[i, :] = stim[spk-prespk:spk+postspk]
        else: 
            allregion[i, :] = np.nan
        print(i, spk)
    sta = np.nanmean(allregion, axis=0)
    return sta
        
sta = get_sta(stim, spks, 70, 10)
```
